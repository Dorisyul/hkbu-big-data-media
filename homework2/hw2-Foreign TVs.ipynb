```python
tvs=[]
```

```python
for mytv in mytvs:
    mynumber=mytv.find("div",attrs={"class":"a1"}).find("strong").text
    mytitle=mytv.find("div",attrs={"class":"a0"}).find("div",attrs={"class","fl info"}).find("a").find("span").text
    mytype=mytv.find("div",attrs={"class":"a0"}).find("div",attrs={"class","fl info"}).find("p").text.split('(')[0]
    mycollection=mytv.find("div",attrs={"class":"a0"}).find("div",attrs={"class","fl info"}).find("p").text.split('(')[1]
    myyear=mytv.find("div",attrs={"class":"a1"}).find("p").text.split(' ')[0]
    mycountry=mytv.find("div",attrs={"class":"a1"}).find("p").text.split(' ')[1]
    tvs.append([mynumber,mytitle,mytype,mycollection,myyear,mycountry])
```

```python
tvs
```

```python
import csv
with open("mytvs.csv","w",newline="") as csvfile:
    spamwriter = csv.writer(csvfile)
    spamwriter.writerows(tvs)
```
